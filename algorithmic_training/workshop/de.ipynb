```python
import random

def make_666():
    count = 1
    test = random.randint(1, 6), random.randint(1, 6), random.randint(1, 6)
    while test != (6, 6, 6):
        test = random.randint(1, 6), random.randint(1, 6), random.randint(1, 6)
        count +=1

    return count, test
```

```python
make_666()
```

```python
import matplotlib.pyplot as plt

made_666 = [make_666()[0] for i in range(1000)]
print(min(made_666), max(made_666))
plt.plot(made_666)
plt.ylabel('Number of tries')
plt.show()
```
